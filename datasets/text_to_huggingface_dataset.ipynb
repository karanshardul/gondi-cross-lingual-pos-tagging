```python
import os
from datasets import Dataset

DATA_DIR = "processed_datasets"
languages = ["hindi", "telugu", "tamil", "marathi", "gondi"]


def read_pos_file(filepath):
    sentences = []
    labels = []

    tokens = []
    tags = []

    with open(filepath, "r", encoding="utf-8") as f:
        for line in f:

            line = line.strip()

            if line == "":
                if tokens:
                    sentences.append(tokens)
                    labels.append(tags)
                    tokens = []
                    tags = []
                continue

            word, pos = line.split("\t")
            tokens.append(word)
            tags.append(pos)

    return {"tokens": sentences, "labels": labels}


datasets = {}

for lang in languages:

    lang_path = os.path.join(DATA_DIR, lang)

    train = read_pos_file(os.path.join(lang_path, "train.txt"))
    dev = read_pos_file(os.path.join(lang_path, "dev.txt"))
    test = read_pos_file(os.path.join(lang_path, "test.txt"))

    datasets[lang] = {
        "train": Dataset.from_dict(train),
        "dev": Dataset.from_dict(dev),
        "test": Dataset.from_dict(test),
    }

print("Datasets loaded successfully")
```
