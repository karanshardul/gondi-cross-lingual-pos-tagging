```python
import os
from conllu import parse_incr

# languages you are using
languages = ["hindi", "telugu", "tamil", "marathi", "gondi"]

# path to datasets
DATA_DIR = "datasets"
OUTPUT_DIR = "processed_datasets"

os.makedirs(OUTPUT_DIR, exist_ok=True)


def preprocess_conllu(file_path, output_file):
    """
    Convert CONLLU file into token POS format
    """

    with open(file_path, "r", encoding="utf-8") as f, open(output_file, "w", encoding="utf-8") as out:

        for tokenlist in parse_incr(f):

            for token in tokenlist:

                # skip multiword tokens
                if isinstance(token["id"], tuple):
                    continue

                word = token["form"]
                pos = token["upos"]

                if word is None or pos is None:
                    continue

                out.write(f"{word}\t{pos}\n")

            # sentence separator
            out.write("\n")


# preprocess all languages
for lang in languages:

    lang_path = os.path.join(DATA_DIR, lang)

    train_file = os.path.join(lang_path, "train.conllu")
    dev_file = os.path.join(lang_path, "dev.conllu")
    test_file = os.path.join(lang_path, "test.conllu")

    output_lang = os.path.join(OUTPUT_DIR, lang)
    os.makedirs(output_lang, exist_ok=True)

    preprocess_conllu(train_file, os.path.join(output_lang, "train.txt"))
    preprocess_conllu(dev_file, os.path.join(output_lang, "dev.txt"))
    preprocess_conllu(test_file, os.path.join(output_lang, "test.txt"))

    print(f"{lang} preprocessing complete")

print("All datasets processed successfully")
```
