```python
import random

pron = ["nang","ni"]
noun = ["mara","illu","manus","penu","pustak","neer"]
verb = ["tin","vel","pad","dek"]
adj = ["bada","chota"]

sentences = []

for _ in range(1000):

    structure=random.choice([
        ["PRON","NOUN","VERB"],
        ["PRON","ADJ","NOUN","VERB"],
        ["NOUN","NOUN","VERB"]
    ])

    sent=[]

    for tag in structure:

        if tag=="PRON":
            word=random.choice(pron)

        elif tag=="NOUN":
            word=random.choice(noun)

        elif tag=="VERB":
            word=random.choice(verb)

        elif tag=="ADJ":
            word=random.choice(adj)

        sent.append((word,tag))

    sentences.append(sent)

with open("gondi.conll","w") as f:

    for sent in sentences:

        for w,t in sent:
            f.write(f"{w} {t}\n")

        f.write("\n")
```
